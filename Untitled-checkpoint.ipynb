```python
#outliers will affect the model so let's create a copy without them
df_no = flights.copy()
df_no = df_no[df_no.DEPARTURE_DELAY > 25]
```

```python
# Remove columns with more than 25% missing values
missing_columns = df_no.loc[:, df_no.isna().mean() >= 0.25].columns
df_no = df_no.drop(missing_columns, axis=1)
df_no.shape
```

```python
# cancelled or diverted flights should not interfere 
df_no = df_no[df_no.CANCELLED != 1]
df_no = df_no[df_no.DIVERTED != 1]
df_no.shape
```

```python
# Drop all columns that do not add interpretability to the model
df_no = df_no.drop(['YEAR','DAY','FLIGHT_NUMBER', 'TAIL_NUMBER','TAXI_OUT','SCHEDULED_DEPARTURE','WHEELS_OFF',
                  'WHEELS_ON','TAXI_IN','SCHEDULED_ARRIVAL','DIVERTED','CANCELLED'], axis=1)
```

```python
# Fill NaN with 0
df_no = df_no.fillna(0)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#dealing with categorical data
```

```python
categorical_cols = df_no.select_dtypes('object').columns.tolist()
df_no[categorical_cols] = df_no[categorical_cols].astype(str)
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore').fit(df_no[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
encoded_cols
```

```python
df_no[encoded_cols] = encoder.transform(df_no[categorical_cols])
```

```python
y = df_no['DEPARTURE_DELAY'].copy()
X = df_no.drop('DEPARTURE_DELAY', axis=1).copy()
```

```python
from sklearn.model_selection import train_test_split
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=48)
```

```python
#let's try multiregression first (no need for scaling when it comes to regression)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create inputs and targets
#inputs, targets = X_train[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY']], y_train
```

```python
lm = LinearRegression()
L_model = lm.fit(X_train,y_train)
L_predictions = pd.DataFrame(lm.predict(X_test))
print(L_predictions.head(50))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
print("MSE =", metrics.mean_squared_error(L_predictions, y_test))
```

```python
DT_accuracy = L_model.score(X_test, y_test)
print('DT Accuracy = '+ str(DT_accuracy.round(2)))
```

```python
from sklearn.tree import DecisionTreeRegressor
DT_regr = DecisionTreeRegressor(max_depth=100, random_state=48)
DT_model = DT_regr.fit(X_train,y_train)
DT_predictions = pd.DataFrame(DT_regr.predict(X_test))
print(DT_predictions.head(50))
print("MSE =", metrics.mean_squared_error(DT_predictions, y_test))
```

```python
DT_accuracy = DT_model.score(X_test, y_test)
print('DT Accuracy = '+ str(DT_accuracy.round(2)))
```
